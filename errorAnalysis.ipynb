```python
# Clean and Unwrangle dataset
columns = ['ticker', 'sector', 'weeklyma', 'monthlyma', 'yearlyma', 'gross/revenue', 'gross/revenue-1yr', 'gross/revenue-2yr', 'net/revenue', 'net/revenue-1yr', 'net/revenue-2yr', 'current-ratio', 'debt-equity', 'debt-asset', 'rating']
df = pd.DataFrame(data = features, columns = columns)
df.set_index('ticker', inplace=True)
df.dropna(inplace=True)
```

```python
# Drop NaN and Null values
df = df[df['sector'] != 'NaN']
df[['sector', 'rating']].groupby('sector').count()
```

```python
# Group and classify the label column for encoding
df[['sector', 'rating']].groupby('rating').count()
```

```python
# Separate features and labels as X and Y
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
# Encode the sector features and labels
X_encode = pd.get_dummies(X.iloc[:,0])
Y_encode = pd.get_dummies(Y)
X = pd.concat([X_encode, X.drop(columns=['sector'])], axis=1, join_axes=[X_encode.index])
```

```python
# Running multiple trial runs to assess accuracy of the model
trial_runs = []
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Running 100 trials, accuracy of the model will vary based randomization of train and test splits
for trial in range(0,100):
    
    # 80/20 Train/Test split
    y = Y_encode.values
    x = X.values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)
    
    # Using the One vs. All Logistic Regression to predict the rating of stocks
    
    model = OneVsRestClassifier(LogisticRegression(solver='liblinear', multi_class='ovr')) # 'lbfgs' , 'liblinear'
    
    # Fit and train model according to training set
    model.fit(x_train, y_train)
    
    # Model predicts the probabilities of each classification
    y_prob = model.predict_proba(x_test)
    
    # Map label to the highest probability
    labels = Y_encode.columns
    y_pred = []
    y_true = []

    for i in y_prob:
        y_pred.append(labels[i.argmax()])

    for i in y_test:
        y_true.append(labels[i.argmax()])
    
    # Get model accuracy
    
    trial_runs.append(round(accuracy_score(y_true, y_pred),2))
```

```python
print(min(trial_runs), max(trial_runs), round(sum(trial_runs)/len(trial_runs), 2))
```

```python
print(classification_report(y_true, y_pred))
```
